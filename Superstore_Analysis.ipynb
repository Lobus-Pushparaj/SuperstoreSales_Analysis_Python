**Hypothesis 1** : Technology products have the highest Sales compared to other categoreis

```python
cat_Sales = data.groupby('Category')['Sales'].sum()
cat_Sales.plot(kind="bar")
plt.title("Sales by categories")
plt.xlabel("categories")
plt.ylabel("Total Sales")
plt.show()
# Conclusion: The Hypothesis supported as Technology products have the highest sales of three categories
```

Hypothesis 2: The East Region has the highest sales compared to other Regions

```python
Reg_Sales = data.groupby('Region')['Sales'].sum()
Reg_Sales.plot(kind='bar')
plt.title("Sales by Region")
plt.xlabel('Region')
plt.ylabel('Sum of Sales')
plt.show()

# Conclusions : The Hypothesis is not supported as the West region has the highest sales compared to other regions
```

Hypothesis 3: Sales are higher during certain Months of the Year

```python
data.columns
```

```python
data['Year'] = pd.DatetimeIndex(data['Order Date']).year
```

```python
Mon_Sales = data.groupby('Month')['Sales'].sum()
Mon_Sales.plot(kind="line")
plt.title("Sales by Month")
plt.xlabel("Month")
plt.ylabel("Sum of Sales")
plt.show()
# Conclusion: This supports our hypothesis that sales are higher during the month of November
```

Hypothesis 4: The company sales is more on weekdays than on weekends

```python
data.columns
```

```python
data['OrderDay'] = pd.DatetimeIndex(data['Order Date']).day_name()
```

```python
data.columns
```

```python
Day_Sales = data.groupby('OrderDay')['Sales'].sum()
Day_Sales.plot(kind="bar")
plt.title("Sales by Day")
plt.xlabel("Days")
plt.ylabel("Sum of Sales")
plt.show()
#Conclusion: The Hypothesis supports as the company sales higher in weekdays than weekends
```
